# Uploading dataset

```python
import pandas as pd
github_link = "https://raw.githubusercontent.com/ADITYA-RAJ1234/DAI-101_Assignment/main/heart_attack_prediction_india.csv"  # Define as a string
df = pd.read_csv(github_link)  # Pass the string to read_csv
print("LOADED.")
```

# UNIVARIATE ANALYSIS

# Summary Statistics

```python
numeric_columns = df.select_dtypes(include=['int64']).columns
summary_stats = df[numeric_columns].describe()
print("Summary Statistics:\n", summary_stats)
```

# HISTOGRAM

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(20, 20))
axes = axes.flatten()

for i, col in enumerate(numeric_columns):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Histogram of {col}')

plt.tight_layout()
plt.show()
```

# BOX PLOT

```python
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(20, 20))
axes = axes.flatten()

for i, col in enumerate(numeric_columns):
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

# BAR CHART

```python
categorical_columns = df.select_dtypes(include=['object']).columns

fig, axes = plt.subplots(nrows=len(categorical_columns), ncols=1, figsize=(10, 5 * len(categorical_columns)))

if len(categorical_columns) == 1:
    axes = [axes]

for i, col in enumerate(categorical_columns):
    sns.countplot(y=df[col], ax=axes[i], order=df[col].value_counts().index)
    axes[i].set_title(f'Bar Chart of {col}')

plt.tight_layout()
plt.show()
```

# AREA CHART

```python
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(20, 20))
axes = axes.flatten()

for i, col in enumerate(numeric_columns):
    df.sort_values(by=col, inplace=True)
    axes[i].fill_between(df.index, df[col], alpha=0.5)
    axes[i].set_title(f'Area Chart of {col}')

plt.tight_layout()
plt.show()
```
